```python
import numpy as np
import pandas as pd
import calendar
data = pd.read_csv('D:\新零售\附件1.csv',encoding='gbk')
data["销量"] = '1'

data.loc[data['地点']=="A",'设备ID']="售货机A"
data.loc[data['地点']=="B",'设备ID']="售货机B"
data.loc[data['地点']=="C",'设备ID']="售货机C"
data.loc[data['地点']=="D",'设备ID']="售货机D"
data.loc[data['地点']=="E",'设备ID']="售货机E"  ##更改设备名称
data.to_csv('D:/新零售/附件0.csv',index=False)  ##index = False禁止生成文件时添加索引列“unnamed”
```

由于每台售货机与A,B,C,D,E五个地点是一一映射关系，所以通过“地点”修改设备名称，以便于查阅。

```python
data0 = pd.read_csv('D:\新零售\附件0.csv')
test1 = data[data["地点"].str.contains('A')]  ##利用contains检查字符串并以此分类
test1.to_csv('D:/新零售/任务1A.csv',index=False,encoding='gbk')

test2 = data[data["地点"].str.contains('B')]
test2.to_csv('D:/新零售/任务1B.csv',index=False,encoding='gbk')

test3 = data[data["地点"].str.contains('C')]
test3.to_csv('D:/新零售/任务1C.csv',index=False,encoding='gbk')

test4 = data[data["地点"].str.contains('D')]
test4.to_csv('D:/新零售/任务1D.csv',index=False,encoding='gbk')

test5 = data[data["地点"].str.contains('E')]
test5.to_csv('D:/新零售/任务1E.csv',index=False,encoding='gbk')

data0["支付时间"]=pd.to_datetime(data0['支付时间'],format='%Y-%m-%d',errors='coerce')
data0 = data0.set_index('支付时间')  ##将“支付时间”设置为索引
data5 = data0['2017-05']
datagb = data5.groupby('设备ID')[['实际金额','销量']].sum()
datagb.loc['sum'] = datagb[['实际金额','销量']].agg(lambda x:x.sum(),axis=0)
datagb.to_csv('D:/新零售/销量金额.csv')
datagb
```

利用index=False阻止python生成文件时在首列添加索引列“unnamed 0”；通过set_index将“支付时间”设定为索引以至于可以直接抽取出2017年5月相关数据

```python
data00 = data0.reset_index()  ##还原索引
data0ct = pd.crosstab(index=data00['设备ID'],columns=data00['支付时间'].apply(lambda m:m.month),values=data00['实际金额'],aggfunc=np.mean)
data0ct.to_csv('D:/新零售/每月每单均值.csv')
data0ct
```

前面我们用set_index将“支付时间”设置为索引，这里我们通过data00 = data0.reset_index()将其还原

```python
datact = pd.crosstab(index=data00['设备ID'],columns=data00['支付时间'].apply(lambda m:m.month),values=data00['销量'],aggfunc=np.sum)
fenmu = calendar.mdays[1:]  ##生成一个包含每月天数的数组，数组第二个元素为1月的天数，所以用[1:]
daymean = datact.div(fenmu,axis=1)  ##series与dataframe相除
daymean.to_csv('D:/新零售/每月日均订单.csv')
daymean
```

datact是存有各个售货机每月销量之和的dataframe，我们导入calendar库，利用calendar.mdays[1:]得到每月天数的Series，在axis=1轴上相除既可。
