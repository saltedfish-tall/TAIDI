```python
import os
from os import path
import numpy as np
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
from PIL import Image
from matplotlib import pyplot as plt
from scipy.misc import imread
import random
import jieba

d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
stopwords = set('')
stopwords.update(['饮料','功能','速食','方便','果蔬',
                  '东鹏','特饮','阿萨姆','奶茶','燕塘',
                  '快线','营养','纯净水','怡宝','夜间',
                  '上午','下午','售出','海味','零食','碳酸'])  ##设置停止词
font_path = 'C:\Windows\Fonts\simhei.ttf'  ##使用本机中文字体


atext = open(path.join(d,r'D:\新零售\词云A.txt'),encoding='UTF-8').read()
atext+=' '.join(jieba.cut(atext,cut_all=False))  ##cut_all=False为精准模式
background_Image = np.array(Image.open(path.join(d, "D:\新零售\A.jpg")))  ##获取图片
img_colors = ImageColorGenerator(background_Image)


wc = WordCloud(
        font_path = font_path,
        max_words = 25,  ##显示的词语的数量
        min_font_size = 4,  ##最小词语的大小
        random_state = 42,
        background_color = 'white',
        mask=background_Image,
        stopwords=stopwords,
        collocations=False  ##禁止使用词语自动搭配
        )
wc.generate_from_text(atext)

wc.to_file('D:\新零售\词云图A饮料.png')

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.show()
```
