```python
import os
from os import path
import numpy as np
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
from PIL import Image
from matplotlib import pyplot as plt
from scipy.misc import imread
import random
import jieba

d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
stopwords = set('')
stopwords.update(['','','','','',
                  '','','','','',
                  '','','','','',
                  '','','','','',''])  ##
font_path = 'C:\Windows\Fonts\simhei.ttf'  ##


atext = open(path.join(d,r'D:\\A.txt'),encoding='UTF-8').read()
atext+=' '.join(jieba.cut(atext,cut_all=False))  ##cut_all=False
background_Image = np.array(Image.open(path.join(d, "D:\\A.jpg")))  ##
img_colors = ImageColorGenerator(background_Image)


wc = WordCloud(
        font_path = font_path,
        max_words = 25,  ##
        min_font_size = 4,  ##
        random_state = 42,
        background_color = 'white',
        mask=background_Image,
        stopwords=stopwords,
        collocations=False  ##
        )
wc.generate_from_text(atext)

wc.to_file('D:\\A.png')

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.show()
```
